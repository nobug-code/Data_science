```python
'''
특정 상황에 가장 적합한 모델을 찾아야 할 때가 많다. 
여기서 '가장 적합한' 이라 함은 대부분 '모델의 오류(error)를 최소하 하는 '
or likelihood(우도)를 최대화 하는' 것을 의미한다. 

어떤, 최적화 문제에 대한 답을 내리는 것이다.

우도란 ?

어떤 시행의 결과 (Evidence) E 가 주어졌다 할 때, 만일 주어진 가설 H 가 참이라면, 
그러한 결과 E 가 나올 정도는 얼마나 되겠느냐 하는 것이다. 즉  결과 E 가 나온 경우,
그러한 결과가 나올 수 있는 여러 가능한 가설들을 평가할 수 있는 측도가 곧 우도인 셈이다.

우도에 대한거 좀 더 자세히 정리하자 

'''
```

```python
'''
경사 하강법
'''

'''
실수 벡터를 입력하면 실수 하나를 출력해 주는 함수 f 가 있다고 가정을 하자
'''
def sum_of_squares(v):
    return sum(v_i ** 2 for v_i in v)
'''

즉, 함수 f를 최대화 시키는 입력값 v 를 찾아야 한다. 

'''
```

```python
'''
gradient(경사, 기울기)
편미분 벡터이고, 나타내는 의미는 함수가 가장 빠르게 증가 할 수 있는 방향을 나타낸다.
'''
'''
함수의 최대값을 구하는 방법 중 하나는 임의의 시작점을 잡은 후, gradient 를 계산하고
gradient의 방향으로 조금 이동하는 과정을 여러 번 반복하는 것이다. 

'''
```

```python
'''
f가 단변수 함수인 경우(변수가  하나 존재하는 것)
점 x 에서의 미분값은 x가 아주 조금 변했을 때 f(x)의 변화량을 의미한다. 
변화량은 h로 나타냄

'''

def difference_quotient(f, x, h):
    return (f(x + h) - f(x))/h
```

```python
import random
def step(v, direction, step_size):
    return [v_i + step_size * direction_i 
           for v_i, direction_i in zip(v, direction)]

def sum_of_squares_gradient(v):
    return [2 * v_i for v_i in v]

def distance(next_v, v):
    
    total = 0 
    for i, j  in zip (next_v, v):
        temp = (i-j) ** 2
        total += temp
    return total
    
v = [random.randint(-10, 10) for i in range(3)]

tolerance = 0.000001

while(True):
    gradient = sum_of_squares_gradient(v)
    next_v = step(v, gradient, -0.01)
    print(next_v)
    if distance(next_v, v) < tolerance:
        break
    v = next_v
```
