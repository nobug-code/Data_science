```python
'''
확률이란 ? 
어떠한 사건의 공간에서 특정 사건이 선택 될 때 발생하는 불확실성을 수치적으로 나타내는 것.
불확실성을 수치적으로 나타낸다는 것이 포인트이다. 

ex) 
주사위를 던져서 1이 나오는 경우 or 주사위를 던져서 짝수가 나오는 경우를 하나의 사건으로 볼 수 있다. 

어떤 사건 E에 대한 확률을 P(E) 라고 하자 

즉 주사위에서 1이 나올 확률은 P(1) = 1/6 일 것이다. 
'''
```

```python
'''
종속 사건
두 사건 E, F 사건 E 가 사건 F 의 발생여부를 제공 한다면 종속 사건
즉 두 사건 E, F 가 연관되어 있는 것 

독립 사건
두 사건 E, F 가 연관이 없는것
ex) 동전을 두 번 던졌을 때 첫번째 동전이 앞면이 나왔더라도 두 번째 동전에서 앞면이 나올 지는 관련이 없다. 

하지만, 첫 번째 동전에서 앞면이 나왔더면 모두 뒷면이 나오는 경우의 발생 여부에 대해서는 알 수  있다. 
이 경우에는 종속 사건이된다. 

독립 사건
P(E, F) = P(E)P(F)
 이다. 
 
조건부 확률
두 사건이 반드시 독립 사건이라는 보장이 없고 사건 F의 확률이 0이 아닌 경우, 사건 E 가 발생할 조건부 확률은 다음과 같다.
P(E|F) = P(E,F)/P(F)

P(E,F) = P(E|F)P(F)

독립 사건일 경우 아래 식이 섭립힌다. 
P(E|F) = P(E)
'''
```

```python
'''
문제

한 가족 내 두 아이들의 성별을 맞추는 예시
다음 두 가지 조건을 가정함
1. 각 아이가 딸이거나 아들일 확률은 동일하다. 
2. 둘째의 성별은 첫째의 성별과 독립이다. 
'''

 '''
두 아이가 모두 딸인 확률은 ? 1/4 이다.
 '''
```

```python
'''
첫째가 딸인 경우(사건 G), 두 아이가 모두 딸일(사건 B)의 확률은 어떻게 될까 ? 
P(B|G) = P(B,G)/P(B)
= 1/2
'''

'''
딸이 최소 한 명인 경우(사건 L), 두 아이가 모두 딸일 확률 또한 계산해 볼 수 있다. 
L = 
P(B|L) = P(B,L)/P(L) = P(B)/P(L) = 1/3
'''
```

```python
'''
컴퓨팅의 장점은 시뮬레이션을 돌려 볼 수 있다는 점에 있다.
'''

import random

def random_kid():
    return random.choice(['boy', 'girl'])

both_girls = 0
older_girl = 0 
either_girl = 0 

random.seed(0)

for _ in range(10000):
    younger = random_kid()
    older = random_kid()
    
    if older == "girl":
        older_girl += 1
    if older == "girl" and younger == "girl":
        both_girls += 1
    if older == "girl" or younger =="girl":
        either_girl += 1
        
print("p(both|older):", both_girls/older_girl)
print("p(both|either):", both_girls/either_girl)
    
```

```python
'''
예시 

P(E|F) = P(E,F)/P(F)

P(E,F) = P(E|F)P(F)

P(T) = 0.99

예를 들어 정확도가 99% 검사기가 있고 , 어떠 사람이 이 검사기로 검사를 시행해서 양성판정이 나왔다면
이 사람이 99% 확률로 병에 걸렸다고 얘기 할 수 있을까? 

이 확률은 병을 가지고 있다는 전제 하에 검사 결과가 양성일 확률이 99%임을 의미한다. 

질병을 걸릴 확률인 P(D) = 1/10,000 = 0.0001 이라고 가정을 하자

            전제              관심사건                 수학적 표현
검사의 정확도  병을 앓고 있다.     검사 결과 : 양성      P(검사결과: 양성 | 병을 앓고 있다.)  = 0.99
우리의 관심사  검사결과 : 양성     병을 앓고 있다.       P(병을 앓고 있다.| 검사 결과 양성)

알고 있는거 P(T|D) = 0.99

P(D|T) = P(T|D)P(D)/P(T)
       = P 0.99 * 0.0001 / P(T) = 0.01
       
P(T) = P(T,D) + P(T,^D)
    = P(T|D)P(D) + P(T|^D)P(^D)
    = 0.99 * 0.0001 + 0.01 * 0.999 

'''
```

```python
'''
연속 분포

동전 던지기는 이산형 분포를 따릅니다. 

이산형 분포 / 이산 그래프에 대해 배웠습니다. 
이산이란 어떤 의미를 나타내는 것이였습니까?


대부분의 경우, 연속적인 결과에 대한 분포를 사용해서 모델을 만듭니다. 

균등 분포는 0과 1 시아의 모든 값에 동등한 비중을 준 분포이다. 

시계의 분침이 1시를 가르칠 확률 -> 0 이다. 


0 과 1 사이에는 무수히 많은 수 가 존재 하기 때문에 숫자 하나의 비중은 0일 것이다. 
이러한 이유 때문에 , 밀도함수를 특정 구간에서 적분한 값으로 확률을 나타내는 확률밀도 함수로 연속분포를 표현한다. 


'''
```

```python
'''
정규 분포 (가우스 분포)

평균값에 많은 데이터가 분포 되어 있고 평균과 떨어 질 수록 데이터가 적게 분포하는 특징을 가지고 있다. 

종혀 곡선 모양의 분포이며, 평균인 u(뮤) 와 표준편차 Q(시그마)의 두 파라미터로 정의됩니다. 
평균은 종의 중심이 어디인지를 나타내며, 표준편차는 종의 폭이 얼마나 넓은지를 나타낸다. 

'''
```

```python
import math

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-mu) ** 2 / 2/ sigma**2) / (sqrt_two_pi * sigma))
```

```python
xs = [x / 10.0 for x in range(-50, 50)]

import plotly.graph_objs as go
fig2 = go.Figure()

fig2.add_trace(
    go.Line(
        x = xs,
        y = [normal_pdf(x, sigma=1) for x in xs],
    #histnorm='percent', percent 를 붙이면 퍼센트로 수치화 되서 나옴
        name='mu=0,sigma=1', # name labels
       #marker_color='#EB89B5',
))

fig2.add_trace(
    go.Line(
        x = xs,
        y = [normal_pdf(x, sigma=2) for x in xs],
    #histnorm='percent', percent 를 붙이면 퍼센트로 수치화 되서 나옴
        name='mu=0,sigma=2', # name labels
       #marker_color='#EB89B5',
))

fig2.add_trace(
    go.Line(
        x = xs,
        y = [normal_pdf(x, sigma=0.5) for x in xs],
    #histnorm='percent', percent 를 붙이면 퍼센트로 수치화 되서 나옴
        name='mu=0,sigma=0.5', # name labels
       #marker_color='#EB89B5',
))

fig2.add_trace(
    go.Line(
        x = xs,
        y = [normal_pdf(x, mu=1) for x in xs],
    #histnorm='percent', percent 를 붙이면 퍼센트로 수치화 되서 나옴
        name='mu=-1,sigma=1', # name labels
       #marker_color='#EB89B5',
))
```
